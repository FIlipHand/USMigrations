```python
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('../data/better_processed/wszystko.csv', sep=';')
data = data.ffill()
data = data.bfill()
data.head()
```

```python
only_data = data.iloc[:, 3:-1]
scaler = StandardScaler()
only_data_scl = pd.DataFrame(scaler.fit_transform(only_data), columns=only_data.columns)
```

```python
column_means = only_data_scl.mean()
deviation_data = (only_data_scl - column_means) / only_data_scl.std()
```

```python
results = deviation_data.filter(like='2020')

results['NPOPCHG2022'] = data['NPOPCHG2022']
results = results.sort_values(by='NPOPCHG2022').tail(15)

results['County'] = data['County']
results['State'] = data['State']
cols = ['State', 'County']
cols.extend(results.columns[:-2])
results = results[cols]
results
```
