```python
import pandas as pd
import numpy as np
import sys, os, pathlib
sys.path.append(str(pathlib.Path(os.getcwd()).parent))
sys.path
from src.utils.projekt_utils import get_project_root
from src.utils.name_utils import US_STATES_DIR
```

```python
core_data = pd.read_csv(os.path.join(get_project_root(), 'data/processed/MigrationData.csv'), encoding='latin-1',
                            sep=';',skipinitialspace = True)

names = ['StateA', 'CountyA','StateB', 'CountyB']
for name in names:
    core_data[name] = core_data[name].map(str.strip)
```

```python
import re
fips = core_data['index'].map(lambda x : re.compile(r'[1-9][0-9]{3}[1-9]?').search(x[1:7]))
```

```python
fips =fips.dropna().map(lambda x: x.group(0))
```

```python
us_data = core_data[core_data['StateA'].isin(US_STATES_DIR.values()) & core_data['StateB'].isin(US_STATES_DIR.values())]
```

```python
us_data
```

```python
import re
fipsA = us_data['index'].map(lambda x : re.compile(r'[1-9][0-9]{3}[1-9]?').search(x[1:7]))
fipsA = fipsA.dropna().map(lambda x: x.group(0))

fipsB = us_data['index'].map(lambda x : re.compile(r'[1-9][0-9]{3}[1-9]?').search(x[7:]))
fipsB = fipsB.dropna().map(lambda x: x.group(0))
```

```python
us_data['FIPS-A'] = fipsA
us_data['FIPS-B'] = fipsB
```

```python
us_data
```

```python
counties = us_data['CountyA'].unique()
counties_dict = dict(zip(counties, range(counties.shape[0])))
```

```python
mat = np.zeros((counties.shape[0], counties.shape[0]))
```

```python
fips = fipsA.unique()
```

```python
fips_dict = dict(zip(fips, range(fips.shape[0])))
```

```python
mat = np.zeros((fips.shape[0], fips.shape[0]))
```

```python
FLOW = 'FlowB-A'
for row in us_data.values:
    # us_data.columns.get_loc('CountyA') == 2
    # us_data.columns.get_loc('CountyA') == 4
    mat[fips_dict[row[us_data.columns.get_loc('FIPS-A')]], fips_dict[row[us_data.columns.get_loc('FIPS-B')]]] = row[us_data.columns.get_loc(FLOW)]
```

```python
mat.shape
```

```python
mat.tofile('../data/processed/migration_matrix.csv',sep=';')
```
